# 카테고리화 해보기

```python
# 데이터 로드
test=pd.read_csv("1908.csv", sep=";")
temp = test.drop_duplicates('기재내용')['기재내용']
# print(t.find('순대'))

# food 카테고리
food = ['순두부','커피','김밥']

# food 카테고리에 포함된 애들 뽑아주기 
for t in temp:
    for f in food:
        if t.find(f)>0:
            print(t)
            
```
